```python
#

def load_tree(path):
  try:
    tree = xml.parse(path)
    return tree, tree.getroot()
  except FileNotFoundError:
    root = xml.Element("library")
    return xml.ElementTree(root), root

def update_price(root, book_id, new_price):
  for book in root.findall("book"):
    if book.get("id") == str(book_id):
      p = book.find("price")
      if p is None:
        p = xml.SubElement(book, "price")
      p.text = str(new_price)
      return True
  return False

def add_book(root, id, title, author, price, category):
  b = xml.SubElement(root, "book", id=str(id))
  xml.SubElement(b, "title").text = title
  xml.SubElement(b, "author").text = author
  xml.SubElement(b, "price").text = str(price)
  xml.SubElement(b, "category").text = category
  return b

def remove_book(root, book_id):
  for book in root.findall("book"):
    if book.get("id") == str(book_id):
      root.remove(book)
      return True
  return False

def save_tree(tree, path):
  tree.write(path, encoding="unicode", xml_declaration=True)

# Example usage:
tree, root = load_tree("library.xml")

# Update price of book id=1 (creates price if missing)
update_price(root, 1, 300)

# Add a new book
add_book(root, 3, "New Book Title", "Some Author", 199, "fiction")

# Remove book id=2
remove_book(root, 2)

# Persist changes
save_tree(tree, "library.xml")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data Simulation

seconds = 48 * 60 * 60   
time = pd.date_range("2024-01-01", periods=seconds, freq="s")

flow = np.zeros(seconds)

# Inject usage bursts
np.random.seed(1)
for _ in range(30):
    start = np.random.randint(0, seconds-300)
    duration = np.random.randint(60, 300)
    flow[start:start+duration] = np.random.uniform(5, 15)

# Inject slow leak for 6 hours
leak_start = 20 * 60 * 60
flow[leak_start:leak_start + 6*60*60] = 0.5

# Add noise during flow
noise = np.random.normal(0, 0.2, seconds)
flow = np.where(flow > 0, flow + noise, flow)

df = pd.DataFrame({"time": time, "flow_lpm": flow})

# Pandas Tasks

# 1. Segment usage events
df["usage"] = df["flow_lpm"] > 1

# 2. Total water usage per hour
df["hour"] = df["time"].dt.hour
hourly_usage = df.groupby("hour")["flow_lpm"].sum()

# 3. Leak detection
df["leak"] = (df["flow_lpm"] > 0.2) & (df["flow_lpm"] < 1)

# 4. Classify usage patterns
def classify(flow):
    if flow < 3:
        return "Low"
    elif flow < 8:
        return "Medium"
    else:
        return "High"

df["usage_type"] = df["flow_lpm"].apply(classify)

# 5. Leakage impact report
leak_volume = df[df["leak"]]["flow_lpm"].sum()

print("Leakage Water Loss (liters):", round(leak_volume,2))
print("\nHourly Usage:\n", hourly_usage)

# Visualization

plt.figure()
plt.plot(df["time"][:20000], df["flow_lpm"][:20000])
plt.title("Water Flow Sensor Sample")
plt.xlabel("Time")
plt.ylabel("Flow (LPM)")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data Simulation

interval = 5
samples = int((24*60*60)/interval)
time = pd.date_range("2024-01-01", periods=samples, freq="5s")

hr = np.ones(samples) * 70

# Sleep (low HR)
hr[0:6*60*12] = 55

# Exercise sessions
for start in [8*60*12, 18*60*12]:
    hr[start:start+30*12] = np.random.uniform(120, 150)
    hr[start+30*12:start+45*12] = np.linspace(150, 80, 15*12)

# Dropouts
drop_idx = np.random.choice(samples, 50)
hr[drop_idx] = np.nan

df = pd.DataFrame({"time": time, "heart_rate": hr})

# Pandas Tasks


# 1. Sleep vs active
df["state"] = np.where(df["heart_rate"] < 60, "Sleep", "Active")

# 2. Detect exercise
df["exercise"] = df["heart_rate"] > 110

# 3. Recovery time
df["recovery"] = (df["heart_rate"] < 90) & (df["heart_rate"].shift(1) > 110)

# 4. Abnormal HR
df["abnormal"] = (df["heart_rate"] > 160) | (df["heart_rate"] < 40)

# 5. Fitness summary
summary = {
    "Avg HR": df["heart_rate"].mean(),
    "Max HR": df["heart_rate"].max(),
    "Sleep Time (hrs)": (df["state"] == "Sleep").sum() * interval / 3600,
    "Exercise Duration (mins)": df["exercise"].sum() * interval / 60
}

summary_df = pd.DataFrame(summary, index=[0])
print(summary_df)

# Visualization

plt.figure()
plt.plot(df["time"], df["heart_rate"])
plt.title("Heart Rate Over 24 Hours")
plt.xlabel("Time")
plt.ylabel("Heart Rate (BPM)")
plt.show()
```
